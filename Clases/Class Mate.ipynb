# Solución de Ecuaciones Diferenciales Parciales mediante Diferencias Finitas

En este notebook, exploraremos la solución numérica de una ecuación diferencial parcial utilizando el método de diferencias finitas. Comenzaremos con el desarrollo de las diferencias finitas para una derivada de primer orden y luego abordaremos la solución de un sistema tridiagonal mediante eliminación gaussiana.

## 1. Desarrollo de Diferencias Finitas

Consideremos una ecuación diferencial parcial con funciones conocidas $a(x)$ y $q(x)$. Después de aplicar el método de diferencias finitas para una derivada de primer orden, obtenemos la siguiente ecuación discreta:

$$
C_1h_{i-1} + C_0h_i + C_2h_{i+1} - Q_i = 0
$$

donde:

$$
\begin{align*}
C_1 &= \frac{1}{2}a(x_i - \frac{\Delta x}{2}) \\
C_2 &= \frac{1}{2}a(x_i + \frac{\Delta x}{2}) \\
C_0 &= -(C_1 + C_2) \\
h_i &= h(x_i) \\
x &\in \{x_{iz}, x_1, x_2, \ldots, x_n, x_{de}\} \\
x_{i+1} &= x_i + \Delta x
\end{align*}
$$

con condiciones de frontera:

$$
h_0 = h_{iz}, \quad h_{n+1} = h_{de}
$$

## 2. Construcción de la Matriz Tridiagonal

Formamos una matriz escalonada con tres diagonales para la primera derivada, más dos extremos por el nivel de derivada. Esta estructura nos lleva a un sistema tridiagonal que puede resolverse eficientemente mediante eliminación gaussiana.

## 3. Método de Thomas (Eliminación Gaussiana para Matrices Tridiagonales)

El método de Thomas es una versión optimizada de la eliminación gaussiana para matrices tridiagonales. A continuación, presentamos el algoritmo en Julia:

````julia
function thomas_algorithm(d0::Vector{Float64}, d1::Vector{Float64}, d2::Vector{Float64}, b::Vector{Float64})
    n = length(b)
    x = zeros(n)
    
    # Eliminación hacia adelante
    for i in 2:n
        α = d1[i] / d0[i-1]
        d0[i] = d0[i] - α * d2[i-1]
        b[i] = b[i] - α * b[i-1]
    end
    
    # Sustitución hacia atrás
    x[n] = b[n] / d0[n]
    for i in (n-1):-1:1
        x[i] = (b[i] - d2[i] * x[i+1]) / d0[i]
    end
    
    return x
end
````

Este algoritmo resuelve eficientemente el sistema $Ah = B$, donde $A$ es una matriz tridiagonal.
4. Solución de la Ecuación con Derivada Temporal
Ahora, consideremos la ecuación completa con una derivada temporal:
$$
G(x)\frac{\partial h}{\partial t} + C_1h_{i-1} + C_0h_i + C_2h_{i+1} - Q_i = 0
$$
Utilizando diferencias hacia adelante para la derivada temporal:
$$
\frac{\partial h(x_i, t_0)}{\partial t} \approx \frac{h(x_i, t_0 + \Delta t) - h(x_i, t_0)}{\Delta t} = \frac{h_i^1 - h_i^0}{\Delta t}
$$
Esto nos lleva a un nuevo sistema tridiagonal con coeficientes modificados:
$$
C_0^* = C_0 - \frac{G}{\Delta t}
$$
Y un nuevo término independiente:
$$
B_i = Q_i - \frac{G_i}{\Delta t}h_i^0
$$
5. Algoritmo de Solución
Aquí presentamos el algoritmo completo para resolver la ecuación, incluyendo la evolución temporal:

````julia
function solve_pde(a::Function, q::Function, g::Function, x::Vector{Float64}, h_iz::Float64, h_de::Float64, Δt::Float64, num_steps::Int)
    n = length(x) - 2  # Número de puntos interiores
    Δx = x[2] - x[1]
    
    # Inicialización
    h = zeros(n+2, num_steps+1)
    h[1, :] .= h_iz
    h[end, :] .= h_de
    
    # Cálculo de coeficientes
    C1 = [0.5 * a(x[i] - 0.5Δx) for i in 2:n+1]
    C2 = [0.5 * a(x[i] + 0.5Δx) for i in 2:n+1]
    Q = [q(x[i]) for i in 2:n+1]
    G = [g(x[i]) for i in 2:n+1]
    
    for t in 1:num_steps
        C0 = -(C1 + C2 + G / Δt)
        B = Q - G / Δt .* h[2:end-1, t]
        
        # Ajuste por condiciones de frontera
        B[1] -= C1[1] * h_iz
        B[end] -= C2[end] * h_de
        
        # Resolver el sistema tridiagonal
        h_new = thomas_algorithm(C0, C2[1:end-1], C1[2:end], B)
        
        # Actualizar la solución
        h[2:end-1, t+1] = h_new
    end
    
    return h
end
````

Este algoritmo resuelve la ecuación diferencial parcial para cada paso de tiempo, actualizando los coeficientes y el término independiente en cada iteración.
6. Ejemplo de Uso
Aquí se muestra cómo usar la función solve_pde:

````julia
# Definir funciones y parámetros
a(x) = 1.0 + 0.5 * sin(π * x)
q(x) = exp(-x^2)
g(x) = 1.0

x = range(0, 1, length=102)
h_iz = 0.0
h_de = 1.0
Δt = 0.01
num_steps = 99

# Resolver la EDP
solution = solve_pde(a, q, g, collect(x), h_iz, h_de, Δt, num_steps)

# Visualizar resultados (requiere el paquete Plots)
using Plots

anim = @animate for t in 1:10:num_steps+1
    plot(x, solution[:, t], xlabel="x", ylabel="h", title="t = $(round((t-1)*Δt, digits=2))")
end

gif(anim, "pde_solution.gif", fps = 5)
````
