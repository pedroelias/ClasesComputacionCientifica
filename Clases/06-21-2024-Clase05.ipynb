# Cálculo iterativo de sen(x) como 4y(1-y)

## Introducción

La función seno es una función trigonométrica fundamental. Existe una forma interesante de aproximar esta función utilizando un método iterativo basado en la ecuación:

```
y_{n+1} = 4y_n(1-y_n)
```

Donde y_n representa la n-ésima iteración del cálculo.

## Proceso iterativo

1. Comenzamos con un valor inicial y_0 = sen^2(x/2).
2. Aplicamos la fórmula iterativa: y_{n+1} = 4y_n(1-y_n).
3. Repetimos el paso 2 varias veces.
4. El resultado final será una aproximación de sen(x).

## Explicación matemática y derivación de la fórmula 4y(1-y)

Para entender cómo llegamos a la fórmula 4y(1-y), necesitamos seguir varios pasos utilizando identidades trigonométricas. Comencemos con algunas identidades clave:

1. $ sen^2(x/2) = (1 - cos(x))/2$
2. $cos(x) = 1 - 2sen^2(x/2)$
3. $sen(x) = 2sen(x/2)cos(x/2)$

Ahora, derivemos la fórmula:

1. Partimos de la identidad $sen(x) = 2sen(x/2)cos(x/2)$

2. Sustituimos cos(x/2) por su expresión en términos de sen(x/2):
   cos(x/2) = √(1 - sen^2(x/2))

3. Ahora tenemos:
   sen(x) = 2sen(x/2)√(1 - sen^2(x/2))

4. Definimos y = sen^2(x/2). Entonces, sen(x/2) = √y

5. Sustituyendo en la ecuación del paso 3:
   sen(x) = 2√y √(1 - y)

6. Simplificando:
   sen(x) = 2√(y(1-y))

7. Elevamos ambos lados al cuadrado:
   sen^2(x) = 4y(1-y)

Esta última ecuación es la base de nuestro método iterativo. Queremos encontrar un valor de y que satisfaga esta ecuación para un sen(x) dado.

8. La iteración y_{n+1} = 4y_n(1-y_n) está diseñada para converger a un punto fijo que satisface y = 4y(1-y), lo cual es equivalente a sen^2(x) = 4y(1-y).

Por lo tanto, al iterar esta fórmula, estamos buscando el valor de y que corresponde a sen^2(x) para un x dado.

## Ejemplo numérico

Supongamos que queremos calcular sen(π/3) ≈ 0.8660254.

1. Comenzamos con y_0 = sen^2(π/6) = 1/4
2. Iteramos:
   - y_1 = 4(1/4)(1-1/4) = 3/4
   - y_2 = 4(3/4)(1-3/4) = 3/4
3. Observamos que el valor converge rápidamente a 3/4.
4. Calculamos sen(π/3) ≈ 2√(3/4) ≈ 0.8660254

## Conclusión

Este método proporciona una forma interesante y computacionalmente eficiente de aproximar la función seno. La fórmula 4y(1-y) surge de una manipulación ingeniosa de identidades trigonométricas, permitiéndonos calcular sen(x) a través de iteraciones simples. Aunque no es tan preciso como otros métodos para todos los valores de x, es notable por su simplicidad y rapidez de convergencia para ciertos ángulos.
