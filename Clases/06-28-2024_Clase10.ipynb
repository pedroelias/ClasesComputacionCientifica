````julia
Base.@kwdef mutable struct Lorenz
    dt::Float64 = 0.02
    ::Float64 = 10
    ::Float64 = 28
    ::Float64 = 8/3
    x::Float64 = 1
    y::Float64 = 1
    z::Float64 = 1
end


function step!(l::Lorenz)
    dx = l. * (l.y - l.x)
    dy = l.x * (l. - l.z) - l.y
    dz = l.x * l.y - l. * l.z
    l.x += l.dt * dx
    l.y += l.dt * dy
    l.z += l.dt * dz
end

atractor=Lorenz()

using Plots

plt = plot3d(1,
    xlim = (-30, 30),
    ylim = (-30, 30),
    zlim = (0, 60),
    title = "Lorenz Attractor",
    legend = false,
    marker = 2,
)

@gif for i in 1:1200
    step!(attractor)
    push!(plt, attractor.x, attractor.y, attractor.z)
end every 10
````

## Introduccin

La computacin paralela es un paradigma que permite la ejecucin simultnea de mltiples instrucciones o tareas para resolver un problema. ## Conceptos Fundamentales

### Procesos vs Threads
- **Procesos**: Instancias independientes de un programa con su propio espacio de memoria.
- **Threads**: Unidades ms ligeras de ejecucin dentro de un proceso, compartiendo el mismo espacio de memoria.

### Granularidad
Se refiere al tamao de las tareas en paralelo:
- **Granularidad fina**: Muchas tareas pequeas.
- **Granularidad gruesa**: Pocas tareas grandes.

### Deadlock
Situacin donde dos o ms procesos se bloquean mutuamente, esperando recursos que el otro tiene.

### Paralelismo de Datos vs Tareas
- **Paralelismo de Datos**: Mismo clculo en diferentes conjuntos de datos.
- **Paralelismo de Tareas**: Diferentes clculos ejecutados simultneamente.

````julia
using Pkg
Pkg.add("BenchmarkTools")
````

````julia
using BenchmarkTools # Necesario instalar BenchmarkTools primero

# Funcin secuencial
function suma_secuencial(a, b)
    return a .+ b
end

# Funcin paralela con Threads
function suma_paralela(a, b)
    result = similar(a)
    Threads.@threads for i in eachindex(a)
        @inbounds result[i] = a[i] + b[i]
    end
    return result
end

# Generar datos
a = rand(1_000_000) #_mejora lectura
b = rand(1_000_000)

# Comparar tiempos
println("Suma secuencial:")
@btime suma_secuencial($a, $b)

println("\nSuma paralela:")
@btime suma_paralela($a, $b)
````

````julia
### 2. Integracin Numrica para Calcular 

# Funcin secuencial

function integralNum(f, ini, fin, num_pasos)
    delta = (fin - ini) / num_pasos
    suma = 0.5 * (f(ini) + f(fin))
    for i in 1:num_pasos-1
        suma += f(ini + i*delta)
    end
    return suma * delta
end

# Funcin paralela mejorada sin Atomic
function integralNum_paralelo_mejorado(f, ini, fin, num_pasos)
    delta = (fin - ini) / num_pasos
    local_sums = zeros(Threads.nthreads())
    
    Threads.@threads for i in 0:num_pasos
        tid = Threads.threadid()
        if i == 0 || i == num_pasos
            local_sums[tid] += 0.5 * f(ini + i*delta)
        else
            local_sums[tid] += f(ini + i*delta)
        end
    end
    
    return sum(local_sums) * delta
end

# Funcin para calcular 
f(x) = 4 / (1 + x^2)

# Clculo y comparacin
resultado_pi = integralNum(f, 0, 1, 1_000_000)
resultado_pi_paralelo = integralNum_paralelo_mejorado(f, 0, 1, 1_000_000)

println(" calculado (secuencial): ", resultado_pi)
println(" calculado (paralelo): ", resultado_pi_paralelo)
println(" en Julia: ", )
println("Error relativo (secuencial): ", abs(resultado_pi - ) / )
println("Error relativo (paralelo): ", abs(resultado_pi_paralelo - ) / )
````

````julia
# Comparacin de tiempos
println("\nTiempo secuencial:")
@btime integralNum($f, 0, 1, 1_000_000)

println("\nTiempo paralelo:")
@btime integralNum_paralelo_mejorado($f, 0, 1, 1_000_000)
````

````julia
### 3. Uso de Threads.@spawn

# Crear 5 arrays aleatorios grandes
tasks = [Threads.@spawn rand(1_000_000) for _ in 1:5]
arrays = [fetch(task) for task in tasks]

function uno(a, b)
    return a .+ b
end

function dos(a, b)
    return a .* b
end

# Versin secuencial
function operaciones_secuenciales(arrays)
    suma = uno(arrays[1], arrays[2])
    mult = dos(arrays[3], arrays[4])
    return suma, mult
end

# Versin paralela
function operaciones_paralelas(arrays)
    suma_task = Threads.@spawn uno(arrays[1], arrays[2])
    mult_task = Threads.@spawn dos(arrays[3], arrays[4])
    return fetch(suma_task), fetch(mult_task)
end

println("Rendimiento secuencial:")
@btime operaciones_secuenciales($arrays)

println("\nRendimiento paralelo:")
@btime operaciones_paralelas($arrays)
````

## Anlisis y Mejores Prcticas

1. **Overhead vs Beneficio**: La paralelizacin introduce overhead. Es beneficiosa cuando el tiempo ahorrado supera este overhead.

2. **Escalabilidad**: Observa cmo cambia el rendimiento al aumentar el nmero de threads y el tamao del problema.

3. **Granularidad**: Tareas muy pequeas pueden no beneficiarse de la paralelizacin debido al overhead.

4. **Uso de Recursos**: Monitorea el uso de CPU y memoria para optimizar el rendimiento.

5. **Sincronizacin**: Minimiza la sincronizacin entre threads para maximizar el paralelismo.

6. **Afinidad de Cache**: Considera la localidad de datos para mejorar el rendimiento en sistemas con mltiples CPUs.

Julia ofrece potentes herramientas para la computacin paralela, desde el uso simple de `Threads.@threads` hasta el ms flexible `Threads.@spawn`. La eleccin entre estos mtodos depende de la naturaleza del problema y la arquitectura del sistema.

La programacin paralela en Julia permite aprovechar al mximo el hardware moderno, pero requiere una comprensin cuidadosa de los conceptos subyacentes y las caractersticas especficas del lenguaje.
