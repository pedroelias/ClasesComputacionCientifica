````julia
Base.@kwdef mutable struct Lorenz
    dt::Float64 = 0.02
    σ::Float64 = 10
    ρ::Float64 = 28
    β::Float64 = 8/3
    x::Float64 = 1
    y::Float64 = 1
    z::Float64 = 1
end


function step!(l::Lorenz)
    dx = l.σ * (l.y - l.x)
    dy = l.x * (l.ρ - l.z) - l.y
    dz = l.x * l.y - l.β * l.z
    l.x += l.dt * dx
    l.y += l.dt * dy
    l.z += l.dt * dz
end

atractor=Lorenz()

using Plots

plt = plot3d(1,
    xlim = (-30, 30),
    ylim = (-30, 30),
    zlim = (0, 60),
    title = "Lorenz Attractor",
    legend = false,
    marker = 2,
)

@gif for i in 1:1200
    step!(attractor)
    push!(plt, attractor.x, attractor.y, attractor.z)
end every 10
````

# Computación Paralela en Julia: Una Guía Completa

## Introducción

La computación paralela es un paradigma que permite la ejecución simultánea de múltiples instrucciones o tareas para resolver un problema. Este enfoque busca mejorar el rendimiento y la eficiencia en el procesamiento de datos, especialmente en problemas complejos o que manejan grandes volúmenes de información. Julia, un lenguaje de programación moderno diseñado para el cálculo científico y numérico de alto rendimiento, ofrece excelentes capacidades para la computación paralela.

## Comparativa de Capacidades de Lenguajes de Programación

## Conceptos Fundamentales

### Procesos vs Threads
- **Procesos**: Instancias independientes de un programa con su propio espacio de memoria.
- **Threads**: Unidades más ligeras de ejecución dentro de un proceso, compartiendo el mismo espacio de memoria.

### Granularidad
Se refiere al tamaño de las tareas en paralelo:
- **Granularidad fina**: Muchas tareas pequeñas.
- **Granularidad gruesa**: Pocas tareas grandes.

### Deadlock
Situación donde dos o más procesos se bloquean mutuamente, esperando recursos que el otro tiene.

### Paralelismo de Datos vs Tareas
- **Paralelismo de Datos**: Mismo cálculo en diferentes conjuntos de datos.
- **Paralelismo de Tareas**: Diferentes cálculos ejecutados simultáneamente.

### Explicación de @time y Distributed

- `@time`: Macro que mide el tiempo de ejecución de una expresión. Proporciona el tiempo total, la memoria asignada y el número de recolecciones de basura.
- `@btime`: Parte del paquete BenchmarkTools, proporciona estadísticas más detalladas y precisas sobre el rendimiento.

## Metodología de Comparaciones de Tiempo

1. **Usar BenchmarkTools**: Proporciona resultados más precisos que `@time`.
2. **Realizar múltiples ejecuciones**: Para obtener estadísticas más robustas.
3. **Calentar la función**: Ejecutar la función una vez antes de medir para evitar la sobrecarga de compilación JIT.
4. **Considerar el tamaño del problema**: La paralelización puede no ser beneficiosa para problemas pequeños debido a la sobrecarga.
5. **Medir la escalabilidad**: Observar cómo cambia el rendimiento al aumentar el número de threads/procesos.

### 1. Suma Secuencial vs Paralela


````julia
using Pkg
Pkg.add("BenchmarkTools")
````

````julia
using BenchmarkTools # Necesario instalar BenchmarkTools primero

# Función secuencial
function suma_secuencial(a, b)
    return a .+ b
end

# Función paralela con Threads
function suma_paralela(a, b)
    result = similar(a)
    Threads.@threads for i in eachindex(a)
        @inbounds result[i] = a[i] + b[i]
    end
    return result
end

# Generar datos
a = rand(1_000_000) #_mejora lectura
b = rand(1_000_000)

# Comparar tiempos
println("Suma secuencial:")
@btime suma_secuencial($a, $b)

println("\nSuma paralela:")
@btime suma_paralela($a, $b)
````

````julia
### 2. Integración Numérica para Calcular π

# Función secuencial

function integralNum(f, ini, fin, num_pasos)
    delta = (fin - ini) / num_pasos
    suma = 0.5 * (f(ini) + f(fin))
    for i in 1:num_pasos-1
        suma += f(ini + i*delta)
    end
    return suma * delta
end

# Función paralela mejorada sin Atomic
function integralNum_paralelo_mejorado(f, ini, fin, num_pasos)
    delta = (fin - ini) / num_pasos
    local_sums = zeros(Threads.nthreads())
    
    Threads.@threads for i in 0:num_pasos
        tid = Threads.threadid()
        if i == 0 || i == num_pasos
            local_sums[tid] += 0.5 * f(ini + i*delta)
        else
            local_sums[tid] += f(ini + i*delta)
        end
    end
    
    return sum(local_sums) * delta
end

# Función para calcular π
f(x) = 4 / (1 + x^2)

# Cálculo y comparación
resultado_pi = integralNum(f, 0, 1, 1_000_000)
resultado_pi_paralelo = integralNum_paralelo_mejorado(f, 0, 1, 1_000_000)

println("π calculado (secuencial): ", resultado_pi)
println("π calculado (paralelo): ", resultado_pi_paralelo)
println("π en Julia: ", π)
println("Error relativo (secuencial): ", abs(resultado_pi - π) / π)
println("Error relativo (paralelo): ", abs(resultado_pi_paralelo - π) / π)
````

````julia
# Comparación de tiempos
println("\nTiempo secuencial:")
@btime integralNum($f, 0, 1, 1_000_000)

println("\nTiempo paralelo:")
@btime integralNum_paralelo_mejorado($f, 0, 1, 1_000_000)
````

````julia
### 3. Uso de Threads.@spawn

# Crear 5 arrays aleatorios grandes
tasks = [Threads.@spawn rand(1_000_000) for _ in 1:5]
arrays = [fetch(task) for task in tasks]

function uno(a, b)
    return a .+ b
end

function dos(a, b)
    return a .* b
end

# Versión secuencial
function operaciones_secuenciales(arrays)
    suma = uno(arrays[1], arrays[2])
    mult = dos(arrays[3], arrays[4])
    return suma, mult
end

# Versión paralela
function operaciones_paralelas(arrays)
    suma_task = Threads.@spawn uno(arrays[1], arrays[2])
    mult_task = Threads.@spawn dos(arrays[3], arrays[4])
    return fetch(suma_task), fetch(mult_task)
end

println("Rendimiento secuencial:")
@btime operaciones_secuenciales($arrays)

println("\nRendimiento paralelo:")
@btime operaciones_paralelas($arrays)
````

## Análisis y Mejores Prácticas

1. **Overhead vs Beneficio**: La paralelización introduce overhead. Es beneficiosa cuando el tiempo ahorrado supera este overhead.

2. **Escalabilidad**: Observa cómo cambia el rendimiento al aumentar el número de threads y el tamaño del problema.

3. **Granularidad**: Tareas muy pequeñas pueden no beneficiarse de la paralelización debido al overhead.

4. **Uso de Recursos**: Monitorea el uso de CPU y memoria para optimizar el rendimiento.

5. **Sincronización**: Minimiza la sincronización entre threads para maximizar el paralelismo.

6. **Afinidad de Cache**: Considera la localidad de datos para mejorar el rendimiento en sistemas con múltiples CPUs.

## Conclusión

Julia ofrece potentes herramientas para la computación paralela, desde el uso simple de `Threads.@threads` hasta el más flexible `Threads.@spawn`. La elección entre estos métodos depende de la naturaleza del problema y la arquitectura del sistema.

La programación paralela en Julia permite aprovechar al máximo el hardware moderno, pero requiere una comprensión cuidadosa de los conceptos subyacentes y las características específicas del lenguaje. Con la práctica y la experimentación, los desarrolladores pueden lograr mejoras significativas en el rendimiento de sus aplicaciones científicas y de cálculo intensivo.
