# Continuación de derivada:

```julia

#=
derivadaNum_vect <- derivada para una funcion discretizada
=#
function derivadaNum_vect(Y,Delta)
    resultado=[]
    for i=1:length(Delta)
        deriv = (Y[i+1]-Y[i])/Delta[i]
        push!(resultado, deriv)
    end
    return resultado
end

function main()
    Mediciones = [8.8, 4.5, 9.7, 4.4, 7.9, 3.4]
    Delta = [1.1, 1.0,0.9,0.99,1.09]
    Derivada=derivadaNum_vect(Mediciones,Delta)
end
@time main()
```

```julia

#=
derivadaNum_vect <- derivada para una funcion discretizada
=#
function derivadaNum_vect(Y,Delta)
    resultado=zeros(length(Delta))
    for i=1:length(Delta)
        deriv = (Y[i+1]-Y[i])/Delta[i]
        resultado[i] = deriv
    end
    return resultado
end

function main()
    Mediciones = [8.8, 4.5, 9.7, 4.4, 7.9, 3.4]
    Delta = [1.1, 1.0,0.9,0.99,1.09]
    Derivada=derivadaNum_vect(Mediciones,Delta)
end
@time main()
```

##  Diferentes esquemas de discretización de la derivada:

### Hacia adelante:

$$
\frac{f_{i-1}-f_i}{\Delta x}

### Hacia atrás:

$$
\frac{f_{i}-f_{i-1}}{\Delta x}

### Mixto:

$$

#  Regresando al problema del conjunto de Mandelbrot:

### Función de multiplicación para  complejos:

$$
z_1=x_1+iy_1\\
z_2=x_2+iy_2\\
\text{Lo que se puede escribir como}\\
Re(z_1\cdot z_2) = x_1x_2-y_1y_2 \\
Im(z_1\cdot z_2) = x_2y_1+x_2y_2\\
$$

```julia
function multiplicacionComplejos(z_1,z_2)
    x_1=z_1[1]
    y_1=z_1[2]
    x_2=z_2[1]
    y_2=z_2[2]
    return (x_1*x_2-y_1*y_2,x_2*y_1+x_1*y_2)
end
```

```julia
@time multiplicacionComplejos([1,1],[2,2])
```

```julia
function multiplicacionComplejos_2(z_1,z_2)
    x_1=z_1[1]
    y_1=z_1[2]
    x_2=z_2[1]
    y_2=z_2[2]
    Real=x_1*x_2-y_1*y_2
    Imaginario=x_2*y_1+x_1*y_2
    return [Real,Imaginario]
end
```

```julia
@time multiplicacionComplejos_2([19,1],[12,3])
```

```julia
using Plots
function multiplicacionComplejos_2(z_1,z_2)
    x_1=z_1[1]
    y_1=z_1[2]
    x_2=z_2[1]
    y_2=z_2[2]
    Real=x_1*x_2-y_1*y_2
    Imaginario=x_2*y_1+x_1*y_2
    return [Real,Imaginario]
end

function modulo(z_1)
    x_1=z_1[1]
    y_1=z_1[2]
    modulo= sqrt(x_1*x_1+y_1*y_1)
    return modulo
end

function DivergenceQ(x)
    if modulo(x)>2
        return true
    else
        return false
    end
end
Dim=1500

Fractal=zeros(Dim,Dim)
x=range(-2,1.5,Dim+1)
y=range(-1.5,1.5,Dim+1)

for i in 1:Dim
    for j in 1:Dim
        a=x[i]
        b=y[j]
        dato=[a,b]
        z=[0,0]
        for k in 1:350
            z=multiplicacionComplejos_2(z,z) + dato
            Fractal[i,j]=modulo(z)
            if DivergenceQ(z) # Q de question.
                break
            end
        end
        #Fractal[i,j]=0
    end
end



heatmap(Fractal, color =:blues)
```
